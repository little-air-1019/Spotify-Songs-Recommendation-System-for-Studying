```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

## Data Cleaning

### Study Playlists

```python
study_data = pd.read_excel("Study Playlist Dataset.xlsx")
study_data.drop(["Unnamed: 0", "track_id"], axis=1, inplace=True)
study_data
```

```python
study_data.drop_duplicates(inplace=True)

duplicates = study_data[study_data.duplicated()]
if duplicates.empty:
    print("No duplicates found in the dataframe.")
else:
    print("Duplicates found in the dataframe:")
    print(duplicates)
```

```python
study_data.dropna(how="any", inplace=True)
```

```python
study_data
```

### Non Study Playlists

```python
non_study_data = pd.read_excel("Non Study Playlist Dataset.xlsx")
non_study_data.drop(["track_id"], axis=1, inplace=True)
non_study_data.dropna(how="any", inplace=True)
non_study_data.drop_duplicates(inplace=True)

duplicates = non_study_data[non_study_data.duplicated()]
if duplicates.empty:
    print("No duplicates found in the dataframe.")
else:
    print("Duplicates found in the dataframe:")
    print(duplicates)

non_study_data
```

### Remove Track if it is in Studyplaylist

```python
study_tracks = study_data["track_name"].values.tolist()
non_study_data = non_study_data[~non_study_data["track_name"].isin(study_tracks)]
non_study_data
```

## Concat the Datasets for Logistic Regression

```python
dataset = pd.concat([study_data, non_study_data])
dataset = dataset.sample(frac=1).reset_index(drop=True)  # Shuffle into random order

dataset
```

```python
dataset.to_excel("Dataset.xlsx")
```
